### Dimensionality Reduction
- PCA(Principal Component Analysis)

### Dimensionality Reduction
- Dimensions are represented as features or columns in dataset.
- Reduction of number of features in our datset
    1. Feature Elimination
    2. Feature Extraction

```python
# import the required packages
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer
```

```python
cancer["data"]
```

```python
df = pd.DataFrame(cancer["data"],columns=['mean radius', 'mean texture', 'mean perimeter', 'mean area',
        'mean smoothness', 'mean compactness', 'mean concavity',
        'mean concave points', 'mean symmetry', 'mean fractal dimension',
        'radius error', 'texture error', 'perimeter error', 'area error',
        'smoothness error', 'compactness error', 'concavity error',
        'concave points error', 'symmetry error',
        'fractal dimension error', 'worst radius', 'worst texture',
        'worst perimeter', 'worst area', 'worst smoothness',
        'worst compactness', 'worst concavity', 'worst concave points',
        'worst symmetry', 'worst fractal dimension'])
df.head()
```

```python
df["target"] = cancer["target"]
df.head()
```

```python
df.shape
```

```python
# checking for null values
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
# extracting features from original datset
x = df.drop("target",axis=1)
```

```python
x.head()
```

```python
# selecting the target
y = df["target"]
```

```python
y.head()
```

```python
# Applying scaling technique
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_transformed = scaler.fit_transform(x)
```

```python
x_transformed
```

```python
s = pd.DataFrame(x_transformed)
```

```python
s.describe()
```

```python
# Apply PCA Model
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca = PCA(n_components=2)
```

```python
newfeatures = pca.fit_transform(x_transformed)
```

```python
newfeatures
```

```python
pca_df = pd.DataFrame(newfeatures,columns=["pca1","pca2"])
pca_df.head()
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```
