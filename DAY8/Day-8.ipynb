## Linear Regression with One feature


- Linear model is sum weighted predicted data to target values
- Y=mx+c
- y = target value/OUtput
- m = slope
- x = Input Value
- c = Coefficient/Intercept


### Slope Formula
- m= (x-xmean)*(y-ymean)/(x-xmean)^2

### Coeffient or intercept formula
- c= ymean-(m*xmean)

## ML steps:
- 1.Get or load the data
- 2.Preprocessing the data
- 3.Define input and output
- 4.Apply the model or algorithm
- 5.pass the data to training and testing
- 6.calculate the score or accuracy score

```python
# get the libraries
import pandas as pd
import numpy as np
```

```python
# 1.get the data
df = pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/Salary_Data.csv")
df.head()
```

```python
# 2.preprocessing the data
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
## 3. Define input and output
x=df[['YearsExperience']]
y=df['Salary']
```

```python
## 4.Apply the model or algorithm
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
## 5.test the data
model.predict([[7.1],[10]])
```

```python
df
```

```python
# 6.identify the score
model.score(x,y)*100
```
